```python
import numpy as np
from matplotlib import pyplot as plt
from math import log10
#%matplotlib inline
import matplotlib.widgets
```

```python
#blackbody function in terms of observed frequency, x
def bb(x, z, Tdust, beta, a):
    # beta value controls steepness of the tail of the RJ bb function
    # define some constants
    # nu0 constant defines the frequency at which the dust opacity is equal to 1
    c,h,k,nu0 = 3.0e8, 6.63e-34, 1.38e-23, 1.5e12
    # convert to rest frame frequeny 
    x = (x*(1+z))
    #convert T_CMB to the redshift and assume this the T_dust
    #Tdust = (T_CMB * (1+z))
    # define the blackbody function
    B = (2 * h * x**3 / c**2) / (np.exp(h * x / (k * Tdust)) - 1)
    return 10**a*(1-np.exp(-(x/nu0)**beta))*B #returns the bb flux density at the specified redshift
```

```python
freq = np.linspace(0.1e-6, 1, 1000)  # frequencys in the range 0.1 microns to 1m, 1000 values
```

```python
z = 2
T_CMB = 2.73 
flux_z2 = []
wavelength = []
for i in freq:
    num = bb(i, z, T_CMB*(1+z), 1, 100)
    lam = i/3e8
    wavelength.append(lam)
    flux_z2.append(num)
#print(flux_z2)
#print(wavelength)
S = []
for j in flux_z2:
    S.append(np.log10(j))
#print(S)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xscale('log')
ax.set_yscale('log')

ax.plot(wavelength, S, 'ro')
plt.show()
```
